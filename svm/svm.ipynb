```python
import numpy as np

class SVM:
    """
    Support Vector Machine (SVM) classifier implemented using batch gradient descent.
    """

    def __init__(self, epochs=500, learning_rate=0.01, C=1.0,lambda_=0.01):
        print("SVM initialized")
        self.epochs = epochs
        self.learning_rate = learning_rate
        self.C = C
        self.w = None
        self.b = None
        self.lambda_ = lambda_

    def initial_weights(self, features):
        print("Initializing weights")
        weights = np.zeros(features)  # Start with zeros for stability
        bias = 0
        return weights, bias

    def hypothesis(self, X):
        return np.dot(X, self.w) + self.b

    def hinge_loss(self, y_pred, Y):
        return np.sum(np.maximum(0, 1 - Y * y_pred))

    def train(self, X, Y):
        print("Training started")
        num_features = X.shape[1]
        self.w, self.b = self.initial_weights(num_features)

        for i in range(self.epochs):
            y_pred = self.hypothesis(X)
            margin = Y * y_pred

            # Compute gradients
            gradient_w = np.zeros_like(self.w)
            gradient_b = 0

            for j in range(len(X)):
                if margin[j] < 1:
                    gradient_w += (-self.C * Y[j] * X[j]) + (2 * self.lambda_ * self.w)  # L2 penalty
                    gradient_b += (-self.C * Y[j])  # No L2 penalty on bias


                # gradient_w += self.w  # Regularization term

            # Average gradients
            gradient_w /= len(X)
            gradient_b /= len(X)

            # Update weights and bias
            self.w -= self.learning_rate * gradient_w
            self.b -= self.learning_rate * gradient_b

            if i % 100 == 0:
                loss = self.hinge_loss(y_pred, Y)
                print(f"Epoch {i}, Loss: {loss}")

        print("Training completed")

    def predict(self, X):
        y_pred = self.hypothesis(X)
        return np.sign(y_pred)
```

```python
s = SVM(epochs=1000,learning_rate=0.1,C=1.0,lambda_=0.01)
s.train(X,Y)
print(s.w)
print(s.b)
```
